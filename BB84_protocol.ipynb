Implementation of BB84 Protocol using polarization.

Group members: Anubhav kumar, Saransh Shivhare, Nathanael Kanter, Happy Pachori

Abstract: We provide 100 % secure communication between two parties using properties of quantum machanics. With the help of BB84 protocol we can make sure that the data is being transmitted to right person. The polarization property of photon is used in communication. We select 2 different basis for transmission based of polarization of photon. Alice (transmitter) selects a random basis to send data and bob (receiver) also selects a random basis for measurement. Then they match basis' from classical channel which will tell them is someone else (eve) listening to their conversation.

In polarization, we use 2 different set of basis for transmission which are |± 45> and |0>,|90> . Their relation can be written as 

$$|± 45> = (1/\sqrt{2}) |0> ± |90> $$

 And vise versa 


encoding to bits into polarization basis can be done as follows; \\
In daigonal basis $-45^\circ$ photon is 0 and  $+45^\circ$ photon is 1 and in horizontal and vertical basis $0^\circ$ photon is 0 and $90^\circ$ photon is 1.

Alice uses 4 lasers to produce 4 different polarization of photons which are $-45^\circ$, $+45^\circ$,$0^\circ$ and $90^\circ$. All of the lasers are focused onto the same channel using beam splitters. Then a filter is used to reduce the average number photons transmitted to $1$. Alice selects the basis randomly and sends the photon in a quantum channel which can optical fiber. Since optical fiber can change the polarization of a photon the waveplates are used to collaborate the channel to neglect the change in polarization by channel. On the Bob side he uses a $50-50$ beam splitter to randomlly choose basis. $Basis$ $1$ in figure is for $± 45^\circ$ polarization and $Basis$ $2$ is for $0^\circ$ sends the respective bit to Eve, but Eve measures the bit on a non-same basis. Now there is a 50-50 percent possibility of Eve getting the actual bit that was sent by Alice. After that, Bob measures the bit sent by Eve on the opposite basis as Alice. When Bob and Alice communicate through the classical channel, they find that they are measuring the bit on a non-same basis, due to which the bit is discarded. and $90^\circ$ degree polarization. To diffrentiate between $0^\circ$ and $90^\circ$ photon we use polarized beam splitter which sends a $0^\circ$  photon straight and and reflects $90^\circ$ photon to the other detector or vice versa. In $Basis 1$ $λ /2$ plate is used to change $45^\circ$ basis to $0^\circ$ and $90^\circ$ basis. At last the Single photon avalance detector is used to detect the photon which detector will detect the photon will also tell the bit of the photon. 

Assuming we have Eve present between communication and also uses same setup to receive Alice bit and same setup as alice to send bit to Bob. after that Alice and Bob can talk on classical channel and discuss basis. Now lets see those cases.

Case - 1 \\
Eve and Bob uses same basis as Alice used. If Eve uses the same basis that Alice used she can detect the correct bit and will send that bit to Bob in the same basis and since Bob is using the same basis he will get a correct bit 100 percent of the time. When Alice and Bob talk about basis they find they use same basis and also got correct bit, so they will keep the bit. In this case Eve is undetected. \\
Ex. If Alice 0 in $0^\circ$ and $90^\circ$ basis which means $0^\circ$ polarized photon and Eve will detect 0 bit in $0^\circ$ and $90^\circ$ basis and will also send $0^\circ$ polarized photon to Bob and Bob will also get 0 bit. So the basis are the same and bit is also the same but Eve also got the data and will be undetected.


Case - 2 \\
Eve will use the same basis as Alice but Bob will use the daigonal basis. In this case Eve will get the correct bit and send the same to Bob but Bob will measure it in different basis. So Bob has a 50-50 chance to getting correct or incorrect basis. When Alice and Bob talk basis they find that they used opposite basis so they discard that recieved bit. Even if Eve got the correct bit the bit is discarded.

Case - 3 \\
Eve uses non same basis but Bob uses the same basis as Alice. Alice sends the respective bit to Eve, but Eve measures the bit on a non-same basis. Now there is a 50-50 percent possibility of Eve getting the actual bit that was sent by Alice. Then Eve sends the bit she recieved to Bob on non same basis as Alice. After that, Bob measures the bit sent by Eve on the same basis as Alice; since Bob measured the bit in an opposite basis to Eve but same to Alice, now Bob has a 50-50 percent possibility of getting the actual bit that Alice sent. Bob and Alice communicate through the classical channel and talk basis and know that they use same basis, so if Bob get the bit send by alice then it's ok but if Bob recieved the other bit then we can say with absolute certainty that Eve is monitoring the transmission, and after detecting Eve, the transmission is stopped. \\
Ex. If Alice send 1 in $0^\circ$ and $90^\circ$ basis which means $90^\circ$ polarized photon, Eve has 50-50- chance of getting 0 or 1. If Eve detects 1, she assumes the photon in $45^\circ$  and send $45^\circ$ photon to Bob. Now Bob measures in $0^\circ$ and $90^\circ$ basis and has a 50-50 percent chance. If Bob gets a 0 bit, after knowing the basis, Alice and Bob are sure that there is eve present.  

Case - 4 \\
Alice sends the respective bit to Eve, but Eve measures the bit on a non-same basis. Now there is a 50-50 percent possibility of Eve getting the actual bit that was sent by Alice. After that, Bob measures the bit sent by Eve on the opposite basis as Alice. When Bob and Alice communicate through the classical channel, they find that they are measuring the bit on a non-same basis, due to which the bit is discarded.

To actually send a secret msg to bob using BB84 we need to send a bunch of bits.Let's say we send 1000 bits through quantum channel. After that alice and bob revieal there choosen basis in classical channel for all 1000 bits but not the bits measured. To know if the communication is secure alice and bob sacrifies few bits in classical channel. So say out of 1000 bit ,200 bits are reviealed in classical channel , then they both see case - 3 . In 200 bit 50 bits are of case - 3 (assuming 50-50 chance of selecting basis by alice). Out of 50 bits , If bob get 50 correct bit send by alice then the communication is 100 percent secure and they have a secret msg out of 800 bit which only they know. Or bob get 25 correct bits and 25 incorrect bits they they know about eve and discard all 1000 bits.
